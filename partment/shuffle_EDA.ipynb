```python
# 将不同类型的col_name打印分类出来。
# 导入 pandas（确保 train_data 已加载）
import pandas as pd

# 数值型列（int、float 等）
numeric_cols = train_data.select_dtypes(include=['int64', 'float64']).columns.tolist()

# 分类型列（object 表示字符串，也可以代表分类数据）
categorical_cols = train_data.select_dtypes(include=['object']).columns.tolist()

# 布尔型列（如果有）
bool_cols = train_data.select_dtypes(include=['bool']).columns.tolist()

# 其他类型列（比如 datetime 类型）
other_cols = train_data.select_dtypes(exclude=['int64', 'float64', 'object', 'bool']).columns.tolist()

# 打印结果
print("数值型列：", numeric_cols)
print("分类型列：", categorical_cols)
print("布尔型列：", bool_cols)
print("其他类型列：", other_cols)
```

```python
# 上图的关系矩阵只是数值型的，像Neighborhood这种离散型数据则没有参与计算，所以下面尝试着使用sklearn来对这些特征进行处理。
import matplotlib.pyplot as plt
from sklearn import preprocessing
import seaborn as sns

f_names=['CentralAir', 'Neighborhood']
for x in f_names:
    label = preprocessing.LabelEncoder()
    train_data[x] = label.fit_transform(train_data[x])
corrmat = train_data.corr()
f, ax = plt.subplots(figsize=(20, 9))
sns.heatmap(corrmat, vmax=0.8, square=True)
```

```python
# 一般认为房子面积越大，售价越高，从图上看大致也是这个趋势。
var ='GrLivArea'
data = pd.concat([train_data['SalePrice'], train_data[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0, 800000), color='blue')
# 右边离群的两个点需要删去，可能表示贫困地区，为异常值
```

```python
# 删除异常值
# 找出异常点
train_data[(train_data['GrLivArea'] > 4000) & (train_data['SalePrice'] < 300000)]
```

```python
# 删除满足条件的异常值
train_data = train_data.drop(train_data[(train_data['GrLivArea'] > 4000) & (train_data['SalePrice'] < 300000)].index)
```

```python
# 再画一次图确认是否删除
var = 'GrLivArea'
data = pd.concat([train_data['SalePrice'], train_data[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0, 800000), color='blue')
```

```python
# 中央空调和SalePrice的关系
var = 'CentralAir'
data = pd.concat([train_data['SalePrice'], train_data[var]], axis=1)
fig=sns.boxplot(x=var, y="SalePrice", data=data)
# fig.axis(ymin=0, ymax=800000)
```

```python
# 整体装修和SalePrice的关系
var = 'OverallQual'
data = pd.concat([train_data['SalePrice'], train_data[var]], axis=1)
fig=sns.boxplot(x=var, y="SalePrice", data=data)
```

```python
# 对比离散数据
"""
sns.hitplot(data,x,y,hue,ax)
data:pandas.Dataframe,numpy.ndarray,mapping,or sequence：input data
x,y : 指定x,y轴的变量
hue：确定绘图颜色的变量
ax:预先定义的绘图区域
"""
f,axes = plt.subplots(3,6,figsize=(30,10),sharex=False)
for i,feature in enumerate(discrete_features):
    sns.histplot(data=combined_df,x=feature,hue='Label',ax=axes[i%3,i//3])          # 可以看到一些特征以0为主，将其删除
```

```python
# 对比连续数据
#对比连续数据
f,axes = plt.subplots(4,6,figsize=(30,15),sharex=False)
for i,feature in enumerate(continuous_features):
    sns.histplot(data=combined_df,x=feature,hue='Label',ax=axes[i%4,i//4])
```

```python
# 检查数值数据的线性分布， 可以发现很多特征关于价格标签并非是线性的。
"""
横轴为数值数据特征，纵轴为价格标签
"""
f,axes = plt.subplots(7,6,figsize=(30,30),sharex=False)
for i,feature in enumerate(numerical_features):
    sns.scatterplot(data=combined_df,x=feature,y="SalePrice",ax=axes[i%7,i//7])
```

```python
# 对比非数值型数据对比分析
f,axes = plt.subplots(7,7,figsize=(30,30),sharex=False)
for i,feature in enumerate(categorical_features):
    sns.countplot(data=combined_df,x=feature,hue="Label",ax=axes[i%7,i//7])
```

```python
# 通过箱状图分析非数值型数据的分布（值取的对应价格）
f, axes = plt.subplots(7,7 , figsize=(30, 30), sharex=False)
for i, feature in enumerate(categorical_features):
    sort_list = sorted(combined_df.groupby(feature)['SalePrice'].median().items(), key= lambda x:x[1], reverse = True)
    order_list = [x[0] for x in sort_list ]
    sns.boxplot(data = combined_df, x = feature, y = 'SalePrice', order=order_list, ax=axes[i%7, i//7])
plt.show()
```

```python
# 检查数值型数据的分布特征并填充均值
import warnings
null_features_numerical = [col for col in combined_df.columns if combined_df[col].isnull().sum()>0 and col not in categorical_features]
plt.figure(figsize=(30,20))
sns.set()
warnings.simplefilter('ignore')
for i,var in enumerate(null_features_numerical):
    plt.subplot(4,3,i+1)
    sns.distplot(combined_df[var],bins=20,kde_kws={'linewidth':3,'color':'red'},label="original")
    sns.distplot(combined_df[var],bins=20,kde_kws={'linewidth':2,'color':'yellow'},label="mean")
```

```python
# # 检查数值型数据的分布特征并填充中位值
plt.figure(figsize=(30,20))
sns.set()
warnings.simplefilter("ignore")
for i,var in enumerate(null_features_numerical):
    plt.subplot(4,3,i+1)
    sns.distplot(combined_df[var],bins=20,kde_kws={'linewidth':3,'color':'red'},label="original")
    sns.distplot(combined_df[var],bins=20,kde_kws={'linewidth':2,'color':'yellow'},label="median")
```

```python
# 时序特征分析
year_feature = [col for col in combined_df.columns if "Yr" in col or 'Year' in col]
year_feature
```

```python
# 然后检查一下这些特征与销售价格是否有关系
combined_df.groupby('YrSold')['SalePrice'].median().plot() # groupby().median()表示取每一组的中位数
plt.xlabel('Year Sold')
plt.ylabel('House Price')
plt.title('House price vs YearSold')
```

```python
# 绘制其他三个特征与销售价格的散点对应图
# 可以看到随着时间的增加，价格是逐增加的
for feature in year_feature:
    if feature != 'YrSold':
        hs = combined_df.copy()
        plt.scatter(hs[feature],hs['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```
